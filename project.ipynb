```python
import requests
```

```python
res = requests.get("https://api.github.com/emojis")
icons=res.json()  # Decode using UTF-8 encoding (common for text)
for name in icons:
    print (name)
```

```python
choosen_icon=input("please choose icon name from the list to display")
```

```python
print(choosen_icon)
```

```python
import urllib.request 
from PIL import Image 
  
```

```python
def display_icon(icon_name):
  urllib.request.urlretrieve(icons[icon_name],"icon.png") 
  img = Image.open("icon.png") 
  img.show()
```

```python
display_icon(choosen_icon)
```
